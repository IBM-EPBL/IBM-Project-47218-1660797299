```python
import matplotlib.pyplot as plt
import numpy as np
from skimage.io import imread
```

```python
A = imread('/content/3a1dbeee-089c-43f0-8f51-a92d3687a515___RS_HL 1754.JPG')
B = imread('/content/1a1184f8-c414-4ead-a4c4-41ae78e29a82___RS_HL 1971.JPG')
```

```python
plt.figure()
plt.subplot(121), plt.imshow(A)
plt.subplot(122), plt.imshow(B)
plt.show()
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(np.abs(A[:, :, 0].astype(float) - B[:, :, 0].astype(float)), cmap='gray')
plt.show()
```
